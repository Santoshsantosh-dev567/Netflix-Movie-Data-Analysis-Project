### **1) What is the most frequent genre of movies released on Netflix?**

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_csv(r"C:\Users\SANTOSH\Downloads\mymoviedb.csv",lineterminator = '\n')
gener_list = df['Genre'].str.split(',')
flattened = []
for sublist in gener_list:
     for gener in sublist:
            flattened.append(gener.strip())

#count the valuse
genercount = Counter(flattened)

#visualization part 

plt.figure(figsize=(10,8))
plt.barh(genercount.keys(),genercount.values(),color="blue",height = 0.9)
plt.title("Genres Counting Graph")
plt.xlabel("Number of Movies")
plt.ylabel("Genres")
plt.grid(axis='x',linestyle="--",alpha=0.5)
plt.show()
```

### **2) What genres has the highest votes?**

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

data = pd.read_csv(r"C:\Users\SANTOSH\Downloads\mymoviedb.csv", lineterminator='\n')

# Loop execution part
dictionary = defaultdict(float)

for idx, row in data.iterrows():
    generlist = row['Genre'].split(',')
    value = row['Vote_Average']   # ✅ FIXED

    for gener in generlist:
        dictionary[gener.strip()] += value

# Plotting part
plt.figure(figsize=(10, 8))
plt.barh(dictionary.keys(), dictionary.values(), color="skyblue")
plt.title("Total Vote Average by Genre")
plt.xlabel("Sum of Vote Average")
plt.ylabel("Genre")
plt.show()

```

### **3) What movie got the highest popularity? What's its genre?**

```python
max_popularity_row = data.loc[data['Popularity'].idxmax()]
print("Highest Popularity Movie:")
print("Title:", max_popularity_row['Title'])
print("Genre:", max_popularity_row['Genre'])
print("Popularity:", max_popularity_row['Popularity'])

```

### **4) What movie got the lowest popularity? What's its genre?**

```python
min_popularity_row = data.loc[data['Popularity'].idxmin()]
print("\nLowest Popularity Movie:")
print("Title:", min_popularity_row['Title'])
print("Genre:", min_popularity_row['Genre'])
print("Popularity:", min_popularity_row['Popularity'])
```

### **5) Which year has the most filmed movies?**

```python
import pandas as pd

# Convert Release_Date to datetime
data['Release_Date'] = pd.to_datetime(data['Release_Date'])

# Extract year from Release_Date
data['Year'] = data['Release_Date'].dt.year

# Find the year with the most movies
most_filmmed_year = data['Year'].value_counts().idxmax()
movie_count = data['Year'].value_counts().max()

print(f"The year with the most filmed movies is {most_filmmed_year}, with {movie_count} movies.")

```
